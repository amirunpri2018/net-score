##Other explorations

###Throughput measurements
I also gathered data on the throughput measured to each of these servers. Most of the results were inconclusive because the Speedtest CLI uses multi-threaded HTTP connections. My methodology, by contrast, was as follows:

1. Run Speedtest-CLI to server and capture reported throughput (download speed)
2. Use Python's HTTLlib to generate a GET request for the same file the CLI is requesting
3. Time how long the request takes to respond, using the wall clock.
4. Divide the length of the response by the measured response time, minus the RTT time, to get a measurement of the speed at which bits are encoded onto the link

Visualizations, more information, and data are available on [my webpage](http://cs.unc.edu/~jamesml). More controls and exploration are needed to discern why throughput varies widely across tools.

    # timer_test() -> Takes a connection and a path to a resource and records times
    # taken for the response
    # args:

    def timer_test(conn, PATH, N):
        r_times = list()
        preIO_times = list()
        for i in xrange(0, N):
            start_wall = time.time()
            resource = "%s?to=%d&i=%d" % (PATH, start_wall, i)

            data = resp.read()

            preIO_times.append(preIO - start_wall)

                print "NOT OK %d (%s)" % (resp.status, resource)

    def server_test(host, path_late, path_thr, to, test_length):
