```python
%matplotlib inline

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/master.csv')
df.head()
```

```python
cols = df.columns.tolist()
#cols
```

```python
throughput = df['ThroughputData.Mean'].values
```

```python
clean = []
#Remove values that do not make sense
for val in throughput:
    if np.isfinite(float(val)) and float(val) > 0:
        clean.append(float(val))
```

```python
clean_max = np.max(clean)
clean_min = np.min(clean)
clean_mean = np.mean(clean)
clean_std = np.std(clean)
```
