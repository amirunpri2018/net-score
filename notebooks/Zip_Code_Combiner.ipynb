```python
zips = pd.read_csv('../data/zip_code_database.csv')
zips.head()
```

```python
cities = zips.primary_city
states = zips.state

print len(cities)
print len(states)
```

```python
zip_dict = {}

for i in xrange(len(cities)):
    city = cities[i].lower()
    state = states[i].lower()
    key = "{0}_{1}".format(city,state)
    zip_dict[key] = zips.zip[i]
```

```python
zip_dict["new york_ny"]
```

```python
df['Zip'] = Series(index=df.index)
df.head()
```

```python
for i in range(len(df['Key'])):
    k = "{0}_{1}".format(df['City'][i].lower(),df['Region'][i].lower())
    if k in zip_dict:
        df['Zip'][i] = zip_dict[k]
    else:
        pass
```
