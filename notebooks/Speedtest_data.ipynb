```python
#array: numpy array
#prec: number of std dev away from mu to remove
def smooth(arr, prec):
    if type(arr) is not list:
        l = arr.tolist()
    else:
        l = arr
    for x in xrange(len(l)):
        l[x] = float(l[x])
    mu = np.mean(l)
    std = np.std(l, dtype=np.float64)
    print "Mu: ", mu
    print "Std: ", std
    
    result = []
    for entry in l:
        if not entry > (mu + (std * prec)) or entry < (mu - (std * prec)):
            result.append(entry)
    return result
```

```python
yt_lat = speedtest_data[:,2]
yt_thru = speedtest_data[:,3]

twc_timer_lat = speedtest_data[:,5]
twc_timer_thru = speedtest_data[:,6]

twc_speed_lat = speedtest_data[:,8]
twc_speed_down = speedtest_data[:,9]

tran_timer_lat = speedtest_data[:,12]
tran_timer_thru = speedtest_data[:,13]

tran_speed_lat = speedtest_data[:,15]
tran_speed_down = speedtest_data[:,16]
```

###Figure 1: Latencies to TWC server 24.25.5.24 via Timer Test (Python HTTPLib) vs. Ookla Speedtest (Speedtest-CLI)
The figure below plots latencies to a TWC server using our timer test (blue) and using the Ookla Speedtest Command Line Interface (red). We notice that the latencies measured by the Ookla Speedtest-CLI are consistently faster than those of the timer test, which makes use of Python's HTTLib.

```python
# Find outlier by converting to a list, then floating value, then finding/removing max. 
# Will need to be rewritten if there are multiple outliers
twc_speed_list = smooth(twc_speed_lat, 1)

pylab.figure()
#pylab.plot(yt_lat)
pylab.plot(twc_timer_lat, '-b')
pylab.plot(twc_speed_list, '-r')
pylab.title('Latencies to TWC server: Timer Test vs Ookla Speedtest-CLI')
pylab.ylabel('Latency')
pylab.xlabel('Test')
```
