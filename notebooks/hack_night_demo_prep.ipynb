##Telling stories
So now we know what the data contains, let's think about what we want to find out... Say my carrier has been really slow on my phone and I don't think I'm getting very good network speeds when using cellular data. I want to shop for a new phone and can change carriers. Let's find out which carrier has the fastest average (mean) network speed!  

First, we have to read the data in. Let's use one of my favorite Python libraries - Pandas.

```python
import pandas as pd
```

```python
df = pd.read_csv('../data_samples/iphone_data_sample.csv')
```

```python
df.info()
df.head()
```

I just want to pause here real quick to explain Pandas DataFrames and Series. DataFrames are pretty much tables and Series are pretty much lists, if you want to read more, the [documentation for pandas](http://pandas.pydata.org/pandas-docs/version/0.15.1/) is very thorough.

Alright, now let's extract out the tests carried out in the United States with fairly straightforward boolean selection. Pandas is super flexible but it is also in-place, so we will have to store the selection as a new DataFrame.

```python
# We want to filter out tests not in the US and the tests that were on WiFi
us_tests = df[ (df['server_country_code'] == 'US') & (df['connection_type'] != 2)]
# We're also going to apply a lambda expression to make all the carrier names lowercase, for later use.
us_tests['carrier_name'] = us_tests['carrier_name'].apply(lambda x: x.lower())

# Keep an eye out for more lambda expressions in this demo, they are quite useful!
```

```python
us_tests.info()
us_tests.head()
```

Now let's slice out the carrier names and find out which carriers were most popular during the test.

```python
carriers = us_tests['carrier_name']
# Yes, easy as that!
```

```python
# Let's make a quick histogram, so we'll pull in a few more libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# These lines adjust the figures that will appear under code cells. We're also resizing the figure to be larger.
%matplotlib inline
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
carrs = {}
for carrier in carriers:
    if carrier not in carrs:
        carrs[carrier] = 0
    carrs[carrier] = carrs[carrier] + 1
```

```python
carrs
```

Now at this point our data set is limited to under 100, so we may not be able to provide key insights. But, with a larger data set, these exact same steps will produce a more meaningful story. Keep that in mind!

```python
plt.bar(range(len(carrs)), carrs.values(), align='center')
plt.xticks(range(len(carrs)), carrs.keys())

plt.show()
```

We can change the colors and add labels, but we'll skip that for now. It looks like the big carriers are represented pretty well, so let's find out the average speed for each.

```python
test = us_tests['download_kbps'].apply(lambda x: np.float64(x)*0.008)
test
```

```python
ofinterest = carrs.keys()

for carrier in ofinterest:
    # Slice out carrier data
    carrier_data = us_tests[us_tests['carrier_name'] == carrier]
    latency = np.asarray(carrier_data['latency'])
    download_speed_mbs = np.asarray(carrier_data['download_kbps'].apply(lambda x: np.float64(x)*0.008))
    # Let's repurpose our old dictionary to contain the information we want.
    count = carrs[carrier]
    carrs[carrier] = [np.mean(latency), np.mean(download_speed_mbs), count]
```

```python
# To show off Pandas' flexibility, let's make a DataFrame out of our new dictionary!
final = pd.DataFrame.from_dict(carrs, orient="index") # Make sure to orient our keys as the index
final.head()
```

```python
# Now let's generate some plots!
latencies = final[0]
speeds = final[1]
plt.bar(range(len(carrs)), latencies, align='center')
plt.xticks(range(len(carrs)), final.index.values)
plt.title('Latencies of Carriers')

plt.show()
```

Okay, so no surprise here. It looks like Movistar and STC (two popular carriers outside the US) have high latencies. Now let's check out what we wanted!

```python
plt.bar(range(len(carrs)), speeds, align='center')
plt.xticks(range(len(carrs)), final.index.values)
plt.title('Download Speeds of Carriers in Mb/s')
plt.xlabel('Carrier')
plt.ylabel('Download speed (Mb/s)')

plt.show()
```
