```python
#array: numpy array
#prec: number of std dev away from mu to remove
def smooth(arr, prec):
    if type(arr) is not list:
        l = arr.tolist()
    else:
        l = arr
    for x in xrange(len(l)):
        l[x] = float(l[x])
    mu = np.mean(l)
    std = np.std(l, dtype=np.float64)
    print "Mu: ", mu
    print "Std: ", std
    
    result = []
    for entry in l:
        if not entry > (mu + (std * prec)) or entry < (mu - (std * prec)):
            result.append(entry)
    return result
```
