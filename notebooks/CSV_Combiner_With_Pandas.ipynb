```python
%matplotlib inline

import pandas as pd
import glob

import numpy as np
```

```python
path = '/playpen/net-score-data/gae_data/2014/'
asn_filenames = glob.glob(path + "/0[0-7]/td_asn*.csv")
loc_filenames = glob.glob(path + "/0[0-7]/tr_loc*.csv")
```

```python
loc_frame = pd.DataFrame()
df_list = []
for loc_file in loc_filenames:
    df = pd.read_csv(loc_file, index_col=['Key','RemoteAddr','GetTime'], header=1)
    df_list.append(df)
loc_frame = pd.concat(df_list)
```

```python
loc_frame.info()
loc_frame.head()
```

```python
asn_frame = pd.DataFrame()
asn_df_list = []
for asn_file in asn_filenames:
    df = pd.read_csv(asn_file, index_col=['Key','RemoteAddr','GetTime'], header=1)
    asn_df_list.append(df)
asn_frame = pd.concat(asn_df_list)
```

```python
asn_frame.info()
asn_frame.head()
```

```python
print "ASN\n", asn_frame.info()
print "\n=========================\n"
print "LOC\n", loc_frame.info()
```

```python
master = pd.merge(asn_frame, loc_frame, left_index=True, right_index=True, how='left', sort=False)
```

```python
master.info()
master.head()
```
